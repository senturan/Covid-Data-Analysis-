### Find maxmimum infection rate for China

```python
covid_agg_data.loc['China'].diff().max()
```

```python
covid_agg_data.loc['Canada'].diff().max()
```

```python
covid_agg_data.loc['Norway'].diff().max()
```

### Find maximum infection rate for all of the countries. 

```python
countries = list(covid_agg_data.index)
max_infection = []
for c in countries:
    max_inf =covid_agg_data.loc[c].diff().max()
    max_infection.append(max_inf)
max_infection    
```

```python
covid_agg_data['max_infection']=max_infection
covid_agg_data
```

### New dataframe with only needed column 

```python
new_covid = pd.DataFrame(covid_agg_data['max_infection'])
new_covid 
```
